```python
import matplotlib.pyplot as plt
import numpy as np

# Data
classifiers = ['ncRNADS', 'Random Forest', 'SVM', 'Logistic Regression', 'K-Nearest Neighbors',
               'Gradient Boosting', 'Decision Tree', 'Nave DRL', 'Neural Network',
               'Extreme Gradient Boosting', 'AdaBoost']

acc = np.array([96.20, 80.12, 81.97, 78.32, 75.68, 82.43, 71.21, 80.26, 85.13, 82.89, 79.45])
prec = np.array([96.48, 80.20, 82.10, 78.40, 75.70, 82.50, 71.20, 82.60, 85.10, 83.00, 79.50])
mcc = np.array([96.10, 62.50, 64.00, 60.20, 58.00, 66.20, 52.50, 63.00, 70.20, 67.50, 61.50])
tpr = np.array([96.29, 80.10, 82.00, 78.30, 75.70, 82.40, 71.20, 80.30, 85.10, 82.90, 79.40])
fpr = np.array([96.20, 19.80, 17.90, 21.60, 24.30, 16.90, 28.40, 18.40, 14.90, 17.10, 20.50])
auc = np.array([96.20, 80.90, 82.10, 78.70, 76.10, 83.00, 71.90, 84.90, 85.60, 83.60, 80.00])
prc_area = np.array([96.48, 78.50, 76.40, 75.20, 71.80, 78.80, 66.70, 83.00, 84.20, 80.20, 76.90])

# Plotting
x = np.arange(len(classifiers))
width = 0.1

fig, ax = plt.subplots(figsize=(14, 8))

# Bar plots
rects1 = ax.bar(x - 2.5*width, acc, width, label='ACC')
rects2 = ax.bar(x - 1.5*width, prec, width, label='PREC')
rects3 = ax.bar(x - 0.5*width, mcc, width, label='MCC')
rects4 = ax.bar(x + 0.5*width, tpr, width, label='TPR')
rects5 = ax.bar(x + 1.5*width, fpr, width, label='FPR')
rects6 = ax.bar(x + 2.5*width, auc, width, label='AUC')
rects7 = ax.bar(x + 3.5*width, prc_area, width, label='PRC Area')

# Labels and Titles
ax.set_xlabel('Classifiers')
ax.set_ylabel('Scores (%)')
ax.set_title('Comparison of Classifier Performance Metrics')
ax.set_xticks(x)
ax.set_xticklabels(classifiers, rotation=45, ha="right")
ax.legend()

# Display plot
plt.tight_layout()
plt.show()
```
